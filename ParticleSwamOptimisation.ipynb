```python
import random
import numpy as np

# 1. Define the Problem (Optimization Function)
# Example: Minimize the function f(x) = x^2
def fitness_function(x):
    return x**2  # This function has a global minimum at x = 0

# 2. Initialize Parameters
num_particles = 30      # Number of particles in the swarm
num_dimensions = 1      # Problem dimensionality (1D for this example)
max_iterations = 100    # Maximum number of iterations
w = 0.5                 # Inertia weight
c1 = 1.5                # Cognitive coefficient
c2 = 1.5                # Social coefficient
x_min, x_max = -10, 10  # Bounds for the position (search space)

# 3. Initialize Particles (Position and Velocity)
class Particle:
    def __init__(self):
        self.position = np.random.uniform(x_min, x_max, num_dimensions)  # Random position
        self.velocity = np.random.uniform(-1, 1, num_dimensions)         # Random velocity
        self.best_position = np.copy(self.position)                      # Best known position
        self.best_fitness = fitness_function(self.position)              # Best known fitness

# Initialize particles
particles = [Particle() for _ in range(num_particles)]

# 4. Evaluate Fitness (already incorporated in the Particle class)

# 5. PSO Main Loop (Update Velocities and Positions)
global_best_position = None
global_best_fitness = float('inf')

# Tracking best solutions
for iteration in range(max_iterations):
    for particle in particles:
        # Update fitness for current particle
        current_fitness = fitness_function(particle.position)

        # Update personal best position if current fitness is better
        if current_fitness < particle.best_fitness:
            particle.best_fitness = current_fitness
            particle.best_position = np.copy(particle.position)

        # Update global best position if current fitness is better
        if current_fitness < global_best_fitness:
            global_best_fitness = current_fitness
            global_best_position = np.copy(particle.position)

    # Update velocities and positions for each particle
    for particle in particles:
        # Inertia term
        inertia = w * particle.velocity

        # Cognitive term (personal experience)
        cognitive = c1 * np.random.random() * (particle.best_position - particle.position)

        # Social term (group experience)
        social = c2 * np.random.random() * (global_best_position - particle.position)

        # New velocity
        particle.velocity = inertia + cognitive + social

        # Update position
        particle.position = particle.position + particle.velocity

        # Ensure position stays within bounds
        particle.position = np.clip(particle.position, x_min, x_max)

    # Print the best fitness at each iteration
    print(f"Iteration {iteration+1}/{max_iterations}, Best Fitness: {global_best_fitness}, Best Position: {global_best_position}")

# 7. Output the Best Solution
print(f"Final Best Position: {global_best_position}, Final Best Fitness: {global_best_fitness}")
```
